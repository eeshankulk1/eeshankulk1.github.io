```python
import altair as alt
import pandas as pd
import os
```

```python
data_url = "https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/licenses_fall2022.csv"
df = pd.read_csv(data_url)
df_smaller = df.sample(n=400, random_state=42)
df_smaller
```

```python

chart1 = alt.Chart(df_smaller).mark_bar().encode(
    x='License Type:N',
    y='count():Q',
    color='License Type:N',
    tooltip=['License Type:N', 'count():Q']
).properties(
    title='Distribution of Categories'
)

chart1
# json_file_path = (r"C:\Users\19259\Desktop\chart.json")
# chart1.save(json_file_path)
# print("success")
```

```python
chart2_base = alt.Chart(df_smaller).mark_bar().encode(
    x='License Status:N',
    y='count():Q',
    color='License Status:N',
    tooltip=['License Status:N', 'count():Q']
).properties(
    title='Distribution of License Status'
)

selection = alt.selection_single(fields=['License Type'], bind='legend')

chart2 = alt.layer(
    chart2_base.add_selection(selection),
    chart2_base.transform_filter(selection)
).properties(
    title='License Status for Selected License Type'
)

chart2
# json_file_path = (r"C:\Users\19259\Desktop\charthw82actual.json")
# chart2.save(json_file_path)
# print("success")
```
